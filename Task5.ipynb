```python
questions=[" 1. How many people have heart disease and how many doesn't have heart disease?",
           " 2. People of which sex has most heart disease?",
           " 3. People of which sex has which type of chest pain most?",
           " 4. People with which chest pain are most pron to have heart disease?"]
questions
```

```python
# lets answer the first question

#1.How many people have heart disease and how many doesn't have heart disease?

#Getting the values
df.target.value_counts( )
```

```python
# Plotting the bar chart
df.target.value_counts( ).plot(kind='bar' , color=['red','green'])
plt.title("Heart Disease Values")
plt.xlabel("1= Heart Disease , 0=No heart Disease")
plt.ylabel("Amount");
```

```python
#plotting a pie chart
df.target.value_counts( ).plot(kind='pie', figsize=(8,6))
plt.legend(["Disease","No Disease"]);
```

```python
# 0 represents 'Female'

# 1 represents 'Male'

# 0 represents 'No Disease'

# 1 represents 'Disease'

df.sex.value_counts( )
```

```python
#plotting a pie chart
df.sex.value_counts( ).plot(kind='pie', figsize=(8,6))
plt.title("Male Female Ratio")
plt.legend(['Male','Female']);
```

```python
# lets find the answer for the second question

#2. People of which sex has most heart disease? 
pd.crosstab(df.target, df.sex)
```

```python
sns.countplot(x='target',data=df,hue='sex')
plt.title("Heart Disease Frequency For Sex")
plt.xlabel("0 = No Heart Disease , 1 = Heart Disease");
```

```python
# Number of Male is double in our dataset than Female.

# More than 45% of male has heart disease and 75% female has heart disease.
```

```python
#Let's move to the third question

#3. People of which sex has which type of chest pain most?

#Counting values for different chest pain
df.cp.value_counts( )
```

```python
#Plotting a bar chart
df.cp.value_counts( ).plot(kind='bar',color=['salmon','lightskyblue','springgreen','khaki'])
plt.title('Chest pain vs count');
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar', color=['coral','lightgreen','black','orange'])
plt.title("Type of chest pain for sex ")
plt.xlabel(" 0 = Female, 1 = Male ");
```

```python
# Most of the male has 'type 0' chest pain and least of male has 'type 4' pain.

# In this case of female 'type 0' and 'type 2' percentage is almosr same.
```

```python
#Now question 4!

#4. People with which chest pain are most pron to have a heart disease?

pd.crosstab(df.cp,df.target)
```

```python
sns.countplot(x= 'cp', data = df, hue = 'target');
```
